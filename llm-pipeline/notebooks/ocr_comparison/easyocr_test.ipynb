# EasyOCR Test

이 노트북은 EasyOCR을 사용하여 한국어/영어 혼용 이미지에서 텍스트를 추출합니다.
결과는 `ocr_results/easyocr/` 폴더에 이미지 이름과 동일한 `.txt` 파일로 저장됩니다.

```python
import os
import easyocr

# 경로 설정
IMAGE_PATH = r'd:\Repos\pro-nlp-finalproject-nlp-01\llm-pipeline\notebooks\extracted_notion_data_v2\images\test2.png'
MODEL_NAME = 'easyocr'
OUTPUT_DIR = os.path.join('..', 'ocr_results', MODEL_NAME)
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Running {MODEL_NAME}...")
reader = easyocr.Reader(['ko', 'en'])
result = reader.readtext(IMAGE_PATH, paragraph=True)
text = "\n".join([res[1] for res in result])

# 결과 저장 (통일된 양식)
img_name = os.path.splitext(os.path.basename(IMAGE_PATH))[0]
output_file = os.path.join(OUTPUT_DIR, f"{img_name}.txt")
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(text)

print(f"{MODEL_NAME} complete. Result saved to {output_file}")
print("--- Extracted Text Preview ---")
print(text[:500] + ('...' if len(text) > 500 else ''))
```
