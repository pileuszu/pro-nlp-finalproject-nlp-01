# Surya OCR Test

```python
import os
import json
import subprocess

# 경로 설정
IMAGE_PATH = r'd:\Repos\pro-nlp-finalproject-nlp-01\llm-pipeline\notebooks\extracted_notion_data_v2\images\test2.png'
MODEL_NAME = 'surya'
OUTPUT_DIR = os.path.join('..', 'ocr_results', MODEL_NAME)
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Running {MODEL_NAME}...")
surya_exe = r'C:\Users\pilla\AppData\Local\Packages\PythonSoftwareFoundation.Python.3.11_qbz5n2kfra8p0\LocalCache\local-packages\Python311\Scripts\surya_ocr.exe'

# CLI 실행 (Surya는 기본적으로 'results/' 폴더를 생성합니다)
cmd = [surya_exe, IMAGE_PATH]

try:
    # 노트북 위치(ocr_comparison)에서 바로 상위의 results 확인을 위해 CWD 설정
    subprocess.run(cmd, check=True, cwd='..')
    
    # 결과 파일 로드 및 .txt 저장
    img_basename = os.path.splitext(os.path.basename(IMAGE_PATH))[0]
    surya_json_path = os.path.join('..', 'results', 'surya', img_basename, 'results.json')
    
    if os.path.exists(surya_json_path):
        with open(surya_json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            lines = data[img_basename][0]['text_lines']
            text = "\n".join([line['text'] for line in lines])
            
            # 통일된 양식으로 결과 저장
            output_file = os.path.join(OUTPUT_DIR, f"{img_basename}.txt")
            with open(output_file, 'w', encoding='utf-8') as f:
                f.write(text)
                
        print(f"{MODEL_NAME} complete. Result saved to {output_file}")
        print("--- Extracted Text Preview ---")
        print(text[:500] + ('...' if len(text) > 500 else ''))
    else:
        print(f"Surya result JSON not found at: {surya_json_path}")
except subprocess.CalledProcessError as e:
    print(f"Surya execution failed with error: {e}")
```
