# Tesseract OCR Test

이 노트북은 Tesseract OCR을 사용하여 이미지를 텍스트로 변환하는 과정을 테스트합니다.
결과는 `ocr_results/tesseract/` 폴더에 이미지 이름과 동일한 `.txt` 파일로 저장됩니다.

```python
import os
import pytesseract
from PIL import Image

# 경로 설정
IMAGE_PATH = r'd:\Repos\pro-nlp-finalproject-nlp-01\llm-pipeline\notebooks\extracted_notion_data_v2\images\2efe77ee-c056-8053-bb95-d3c24330a1b1_2efe77ee-c056-80a4-af65-d2f693417dc8.png'
MODEL_NAME = 'tesseract'
OUTPUT_DIR = os.path.join('..', 'ocr_results', MODEL_NAME)
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Tesseract 실행 파일 및 데이터 자동 설정
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
os.environ['TESSDATA_PREFIX'] = r'd:\Repos\pro-nlp-finalproject-nlp-01\llm-pipeline\notebooks\tessdata'

print(f"Running {MODEL_NAME}...")
custom_config = r'--psm 6 -l kor+eng'
text = pytesseract.image_to_string(Image.open(IMAGE_PATH), config=custom_config)

# 결과 저장 (통일된 양식)
img_name = os.path.splitext(os.path.basename(IMAGE_PATH))[0]
output_file = os.path.join(OUTPUT_DIR, f"{img_name}.txt")
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(text)

print(f"{MODEL_NAME} complete. Result saved to {output_file}")
print("--- Extracted Text Preview ---")
print(text[:500] + ('...' if len(text) > 500 else ''))
```
